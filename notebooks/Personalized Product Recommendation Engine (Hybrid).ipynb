```python
def get_cf_recommendations(username, model, df_cf, n=10):
    all_items = df_cf['id'].unique()
    rated_items = df_cf[df_cf['reviews.username'] == username]['id'].values

    items_to_predict = [item for item in all_items if item not in rated_items]

    predictions = [
        (item, model.predict(username, item).est)
        for item in items_to_predict
    ]

    predictions.sort(key=lambda x: x[1], reverse=True)
    return predictions[:n]
```

Goal
Recommend products similar in content to what a user interacted with.

Input

Product text (review + title)

User interaction history (optional)

Output

Similar products list

```python
df_cb = df.dropna(subset=['reviews.text']).copy()
```

```python
df_cb.columns
```

```python
df_cb['text'] = df_cb['reviews.title'].fillna('') + " " + df_cb['reviews.text']
```

```python
import re

def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-z ]', ' ', text)
    return re.sub(r'\s+', ' ', text).strip()

df_cb['clean_text'] = df_cb['text'].apply(clean_text)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(
    stop_words='english',
    max_features=5000
)

tfidf_matrix = vectorizer.fit_transform(df_cb['clean_text'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity_matrix = cosine_similarity(tfidf_matrix)
```

```python
def get_similar_products(product_id, df_cb, similarity_matrix, n=10):
    if product_id not in df_cb['id'].values:
        return []

    idx = df_cb[df_cb['id'] == product_id].index[0]
    scores = list(enumerate(similarity_matrix[idx]))
    scores = sorted(scores, key=lambda x: x[1], reverse=True)

    top_indices = [i[0] for i in scores[1:n+1]]
    return df_cb.iloc[top_indices]['id'].values

```

```python
def hybrid_recommendations(
    username,
    model_cf,
    df_cf,
    df_cb,
    similarity_matrix,
    n=10,
    cf_weight=0.6,
    cbf_weight=0.4
):
    final_scores = {}

    # CF part
    try:
        cf_recs = get_cf_recommendations(username, model_cf, df_cf, n=20)
        for item, score in cf_recs:
            final_scores[item] = final_scores.get(item, 0) + cf_weight * score
    except:
        pass

    # CBF part (use recent interaction)
    user_history = df_cf[df_cf['reviews.username'] == username]['id']

    if len(user_history) > 0:
        last_item = user_history.iloc[-1]
        cbf_items = get_similar_products(
            last_item, df_cb, similarity_matrix, n=20
        )

        for item in cbf_items:
            final_scores[item] = final_scores.get(item, 0) + cbf_weight

    # Sort & return
    return sorted(final_scores.items(), key=lambda x: x[1], reverse=True)[:n]
```

```python
sample_user = df_cf['reviews.username'].iloc[0]

recs = hybrid_recommendations(
    sample_user,
    model_cf,
    df_cf,
    df_cb,
    similarity_matrix,
    n=10
)

print("Hybrid Recommendations:")
for item, score in recs:
    print(item, round(score, 3))
```
